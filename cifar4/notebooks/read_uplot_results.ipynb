```python
%pylab inline

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager
import seaborn as sns

import sys
sys.path.append('../../code/scripts')
import utils
import plotting as p
from dataset_params import dataset_params

from importlib import reload


```

```python
group_id_dict = {0:'vehicle', 1:'animal'}
group_names = list(group_id_dict.values())

gamma0 = dataset_params['cifar4']['gamma']
gammas = [gamma0, 1- gamma0]
reload(utils)
```

```python
r_ERM = utils.read_in_results(group_key='animal',
                               results_type='subset',
                               results_identifier='cifar4_subsetting_< experiment name >_ERM',
                               obj='ERM',
                               sgd_params = {'lr': 0.001, 'weight_decay': 0.0001, 'momentum': 0.9},
                               num_seeds = 10)

r_IS = utils.read_in_results(group_key='animal',
                             results_type='subset',
                             results_identifier='cifar4_subsetting_< experiment name >_IS',
                             sgd_params = {'lr': 0.001, 'weight_decay':'0.01_0.001', 'momentum': 0.9},
                             obj='IS',
                             num_seeds = 10)


r_GDRO = utils.read_in_results(group_key='animal',
                             results_type='subset',
                             results_identifier='cifar4_subsetting_< experiment name >_GDRO',
                             sgd_params = {'lr': 0.001, 'weight_decay': 0.001, 'momentum': 0.9},
                             obj='GDRO_group_adj_4.0_gdro_stepsize_0.01',
                             num_epochs = '20',
                             num_seeds = 10)

# add 1 - to all
acc_keys = ['auc_roc', 'acc']
for r in [r_ERM, r_IS, r_GDRO]:
    for acc_key in acc_keys:
        # accs_total --add new keys
        new_key  = '1 - {0}'.format(acc_key)
        r[2][new_key] = 1 - r[2][acc_key]
        # accs_per_group
        r[3][new_key] = 1 - r[3][acc_key]
```

# Eval by group.

```python
# preload font sizes, etc.
reload(p)
fig, ax = p.setup_uplot_ax()


ls = ['-','--', ':']
obj_names = ['ERM','IS', 'GDRO'] 
acc_key_plot = '1 - acc'

results_to_plot = [r_ERM, r_IS, r_GDRO]

for i, results in enumerate(results_to_plot):
    groups, subset_sizes, accs_total, accs_by_group = results

    subset_fracs = subset_sizes / subset_sizes.sum(axis=0)
    accs_total_avgs = {}

    p.plot_by_group(accs_by_group,
                    accs_total,
                    subset_fracs,
                    acc_key_plot, 
                    group_id_dict,
                    gammas = gammas,
                    pop_weights = 'by_gamma',
                    ls = ls[i],
                    label_append = obj_names[i],
                    range_type='stddev',
                    title='CIFAR-4 air/land prediction',
                    ylim=None,
                    plot_alpha_star=True, 
                    plot_gamma = True, 
                    gamma_annot_offset = (-0.24,0.09),
                    ax = ax)
    
plt.ylabel('0/1 loss')
#ax.set_yscale('log')#loglog()
#ax.grid(linewidth=1)
ax.set_ylim(0,0.16);

plt.savefig('../../figures/uplot_cifar_acc_90_10.pdf', bbox_inches='tight')
```
